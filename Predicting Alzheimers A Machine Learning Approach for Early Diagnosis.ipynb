### Statistical Analysis

### T-Test

```python
# T-Test for Age and Diagnosis
group1_age = df[df['Diagnosis'] == 0]['Age']
group2_age = df[df['Diagnosis'] == 1]['Age']
t_stat_age, p_val_age = ttest_ind(group1_age, group2_age)
print(f"T-Test for Age and Diagnosis: p-value = {p_val_age}")

# T-Test for BMI and Diagnosis
group1_bmi = df[df['Diagnosis'] == 0]['BMI']
group2_bmi = df[df['Diagnosis'] == 1]['BMI']
t_stat_bmi, p_val_bmi = ttest_ind(group1_bmi, group2_bmi)
print(f"T-Test for BMI and Diagnosis: p-value = {p_val_bmi}")
```

### Chi-Square Test for Categorical Variables

```python
contingency_table_gender = pd.crosstab(df['Gender'], df['Diagnosis'])
chi2_gender, p_gender, dof_gender, ex_gender = chi2_contingency(contingency_table_gender)
print(f"Chi-Square Test for Gender and Diagnosis: p-value = {p_gender}")

# Chi-Square Test for Ethnicity and Diagnosis
contingency_table_ethnicity = pd.crosstab(df['Ethnicity'], df['Diagnosis'])
chi2_ethnicity, p_ethnicity, dof_ethnicity, ex_ethnicity = chi2_contingency(contingency_table_ethnicity)
print(f"Chi-Square Test for Ethnicity and Diagnosis: p-value = {p_ethnicity}")

# Chi-Square Test for EducationLevel and Diagnosis
contingency_table_education = pd.crosstab(df['EducationLevel'], df['Diagnosis'])
chi2_education, p_education, dof_education, ex_education = chi2_contingency(contingency_table_education)
print(f"Chi-Square Test for EducationLevel and Diagnosis: p-value = {p_education}")
```

### Model Developemnt 

```python
# Basline model 
# Compute the mean of the target variable from the training set
mean_diagnosis = y_train.mean()

# Predict the mean for all instances in the test set
y_pred = [mean_diagnosis] * len(y_test)

# Evaluate the mean model
# For binary classification, you might want to threshold the mean to get 0 or 1 predictions
y_pred_class = [1 if pred >= 0.5 else 0 for pred in y_pred]

accuracy = accuracy_score(y_test, y_pred_class)
mse = mean_squared_error(y_test, y_pred)

print(f'Mean model accuracy: {accuracy:.4f}')
print(f'Mean model mean squared error: {mse:.4f}')
```

```python
# Compute the mean of the target variable from the training set
mean_diagnosis = y_train.mean()

# Predict the mean for all instances in the training set
y_train_pred = [mean_diagnosis] * len(y_train)

# Predict the mean for all instances in the test set
y_test_pred = [mean_diagnosis] * len(y_test)

# For binary classification, you might want to threshold the mean to get 0 or 1 predictions
y_train_pred_class = [1 if pred >= 0.5 else 0 for pred in y_train_pred]
y_test_pred_class = [1 if pred >= 0.5 else 0 for pred in y_test_pred]

# Evaluate the mean model on the training set
train_accuracy = accuracy_score(y_train, y_train_pred_class)
train_mse = mean_squared_error(y_train, y_train_pred)

# Evaluate the mean model on the test set
test_accuracy = accuracy_score(y_test, y_test_pred_class)
test_mse = mean_squared_error(y_test, y_test_pred)

print(f'Mean model train accuracy: {train_accuracy:.4f}')
print(f'Mean model train mean squared error: {train_mse:.4f}')
print(f'Mean model test accuracy: {test_accuracy:.4f}')
print(f'Mean model test mean squared error: {test_mse:.4f}')
```

```python
# Initialize models
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree Classifier': DecisionTreeClassifier(),
    'Random Forest Classifier': RandomForestClassifier(),
    'Support Vector Machine': SVC(),
    'Gradient Boosting Classifier': GradientBoostingClassifier()
}
```

```python
# Train and evaluate models
results = {}
for name, model in models.items():
    # Perform 5-fold cross-validation and compute mean accuracy
    cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='accuracy')
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)
    
    results[name] = {
        'Cross-Validation Accuracy': cv_scores.mean(),
        'Accuracy': accuracy,
        'Confusion Matrix': conf_matrix,
        'Classification Report': class_report
    }

```

```python
# Print results
for name, metrics in results.items():
    print(f"Model: {name}")
    print(f"Cross-Validation Accuracy: {metrics['Cross-Validation Accuracy']:.4f}")
    print(f"Accuracy: {metrics['Accuracy']:.4f}")
    print(f"Confusion Matrix:\n{metrics['Confusion Matrix']}")
    print(f"Classification Report:\n{metrics['Classification Report']}")
    print("-" * 40)
```

#### Separetly for Train and Test 

```python
# Define the models
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree Classifier': DecisionTreeClassifier(),
    'Random Forest Classifier': RandomForestClassifier(),
    'Support Vector Machine': SVC(),
    'Gradient Boosting Classifier': GradientBoostingClassifier()
}

# Dictionary to store the results
results = {
    'Model': [],
    'Train Accuracy': [],
    'Test Accuracy': []
}

# Train, predict and evaluate each model
for model_name, model in models.items():
    model.fit(X_train, y_train)
    
    # Predict on train data
    pred_train = model.predict(X_train)
    train_score = accuracy_score(y_train, pred_train) * 100
    
    # Predict on test data
    pred_test = model.predict(X_test)
    test_score = accuracy_score(y_test, pred_test) * 100
    
    # Store the results
    results['Model'].append(model_name)
    results['Train Accuracy'].append(train_score)
    results['Test Accuracy'].append(test_score)
    
    print(f"{model_name} Train Accuracy: {train_score:.2f}%")
    print(f"{model_name} Test Accuracy: {test_score:.2f}%")
    print('----------------------------------------')

# Create a DataFrame for the results
results_df = pd.DataFrame(results)

print(results_df)
```

#### Hyperparameter tuning for RandomForestClassifier

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score

# Define a smaller parameter grid for faster computation
param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 4]
}

# Create a GradientBoostingClassifier object
gb = GradientBoostingClassifier()

# Create a RandomizedSearchCV object for faster hyperparameter tuning
random_search = RandomizedSearchCV(
    estimator=gb,
    param_distributions=param_grid,
    n_iter=10,  # Number of parameter settings to sample
    cv=5,       # Number of cross-validation folds
    scoring='accuracy',
    n_jobs=-1,  # Use all available cores
    verbose=1,  # Set verbosity to 1 for less detailed output
    random_state=42
)

# Fit the random search to the data
random_search.fit(X_train, y_train)

# Print the best parameters and the best score
print(f"Best parameters for Gradient Boosting: {random_search.best_params_}")
print(f"Best cross-validation accuracy: {random_search.best_score_:.4f}")

# Train the GradientBoostingClassifier with the best parameters
best_gb = random_search.best_estimator_

# Predict on train data
pred_train_gb = best_gb.predict(X_train)
gb_train_score = accuracy_score(y_train, pred_train_gb) * 100
print('Tuned Gradient Boosting Train Accuracy: {:.2f}%'.format(gb_train_score))

# Predict on test data
pred_test_gb = best_gb.predict(X_test)
gb_test_score = accuracy_score(y_test, pred_test_gb) * 100
print('Tuned Gradient Boosting Test Accuracy: {:.2f}%'.format(gb_test_score))
```
